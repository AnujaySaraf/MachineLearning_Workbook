```python
# Apriori - Support - A Prior Knowledge

#Support , Confidence, Lift

# Movie-Recommendation - KingArthur
# Support(M)= # Number of People Watched the movie (Troy) / # User Watchlist
# out of team 100 -- 10 people watched  KingArthur = 10%

# Movie-Recommendation - Troy
# Support(M)= # Number of People Watched the movie (Troy) / # User Watchlist
# out of team 100 -- 40 people watched Troy = 40%

# Movie-Recommendation
# Confidence (M1 -> M2)= # user watchlists contains Troy and King Arthur /# Number of People watchlist contains King Arthur 
# out of team 100 people watched Troy 7 people watched Troy King Arthur = 7/40 = 17.5%

# Lift
#Confidence / Lift
# Now letz recommend People watched Troy to watch kingArthur
# Lift = 17.5%/10% = 1.75



# Step1 - Set a minimum and confidence
# Step2 - Take all the sibsets in transactions having higher support than minimum support
# Step3 - Take all rules of these subsets having higher confidence than minimum confidence
# Step4 - Sort the rules by decreasing lift

```

```python
# Import Libraries

import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv')
dataset.head()
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
dataset.head()
```

```python
# Identify the shape of the dataset
dataset.shape
```

```python
# Currently the transaction is as Dataframe , we need to convert this to List to make use of apyori library
# Intialize the empty list transactions
transactions = []

# Scan all the transacations and i is index, j specific transaction and make it as list
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
print ((type(transactions)))
```

```python
#Customer 1
transactions[0]
```

```python
##Customer 2
transactions[1]
```

```python
# Training Apriori on the dataset
# Transaction variable is transaction list
# min_support - The minimum Support of relation
# min_confidence - The minumum confidence of relation
# min_lift -0 The minimum lift of relation
# min_lengh
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
# Visualising the results
results = list(rules)
```
