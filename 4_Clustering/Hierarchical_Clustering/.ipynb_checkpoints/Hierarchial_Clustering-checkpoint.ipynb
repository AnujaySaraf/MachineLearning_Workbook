```python
# Hierarchial Clustering
#Agglomerative and Divisive

# Procedure
# Step1 - Make each datapoint a single-point cluster - This forms N Clusters
# Step2 - Take the two closest data points and make them one cluster - This form N-1
# Step3 - Take the two closest clusters and make them one cluster - This forms N-2
# Step4 - Repeat step 3 until there is one cluster

# Success Model is ready

## Distance between two clusters

#1) Closest Points
#2) Furthest Points
#3) Average Distance
#4) Distance between Centroids

### How Dendogram works
#https://www.gigawiz.com/imagesng4/Dendrogram1.png
#https://dashee87.github.io/images/hierarch_1.gif
```

```python
# Import Libraries

import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset.head()
```

```python
# Identify the shape of the dataset
dataset.shape
```

```python
# Set the Independent Variable (X) and dependent variable (y) = f(x)=y
## Make sure always X is matrix and Y is vector
X = dataset.iloc[:,[3,4]].values
```

```python
# Using the distance (Dendrogram) method to find the optimal number of clusters using scipy
#https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html
#import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import dendrogram,linkage
#dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
Z = linkage(X,method='ward')
dn = dendrogram(Z)
#Z = linkage(X, 'ward')
#dn = dendrogram(Z)

fig = plt.figure(figsize=(25, 10))
plt.title('Dendrogram')
plt.xlabel('Customer Map')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# Fitting Hierarchial Clustering to the dataset
# #http://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html#sklearn.cluster.AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage ='ward')
# ward will minimize the distance between the cluster
y_hc = hc.fit_predict(X)
```

```python
y_hc, len(y_hc)
```

```python
# Visualization the Hierarchial Clusters Results
plt.figure(figsize=(20,10))
# plt.scatter(x,y)
# We need to Plot X axis for annual income and Y axis for Individual Spending Rating
plt.scatter(X[y_hc== 0,0], X[y_hc ==0, 1], s = 100, c='red', label='Careful')
plt.scatter(X[y_hc == 1,0], X[y_hc ==1, 1], s = 100, c='blue', label='Standard')
plt.scatter(X[y_hc == 2,0], X[y_hc ==2, 1], s = 100, c='green', label='Target')
plt.scatter(X[y_hc == 3,0], X[y_hc ==3, 1], s = 100, c='cyan', label='Careless')
plt.scatter(X[y_hc == 4,0], X[y_hc ==4, 1], s = 100, c='magenta', label='Sensible')


plt.title("Clusters of Clients")
plt.xlabel("Average Income in $")
plt.ylabel("Spending Rating 1-100")
plt.legend()
plt.show()
```
