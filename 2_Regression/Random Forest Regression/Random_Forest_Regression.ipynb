```python
# Random_Forest Regression is Non-Linear and Non-continuous Model
# Its team of decision tree and average of each results
# Part of Ensemble Learning 

#Step1 : Pick at Random K DataPoint from the Training Set
# Step 2: Build a decision tree based on those K dataPoints
# Step3 : Choose the number NTree of trees you want to build and repeat STEPS1 & Step2
# Step 3: For a new datapoint make each on you NTree Predict the value of Y to for the datapoint in question.
# assign the new datapoint the average accross all the predicted Y values
```

```python
# Import Libraries


import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset.head()
```

```python
# Identify the shape of the dataset
dataset.info()
```

```python
# Check if we have any null values
dataset.isnull().sum()
```

```python
# Set the Independent Variable (X) and dependent variable (y) = f(x)=y
## Make sure always X is matrix and Y is vector
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
# Fitting the Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=300,random_state=0)
regressor.fit(X,y)
```

```python
Level_Value = 6.5
y_pred = regressor.predict(Level_Value)
print ("The Prediction of {} Level salary is ${}".format(Level_Value,np.round(y_pred)))
```

```python
# Visualising the Regression Results
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue') # Use the Linear regression predict method to Predict the X
plt.title("Truth or Bluff - Random Forest Regression Model")
plt.xlabel("Position_Level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualising the Regression Results for more interval and smoothness in X axis say (0.1) value
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1)) # This help 90 level of salary
plt.figure(figsize=(20,10))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue') # Use the Linear regression predict method to Predict the X
plt.title("Truth or Bluff - Random Forest regression Model")
plt.xlabel("Position_Level")
plt.ylabel("Salary")
plt.show()
```
